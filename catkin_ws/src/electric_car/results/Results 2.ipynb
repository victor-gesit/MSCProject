```python
df_trap = pd.read_csv('trapezoid-1694865717.4499276.csv')

print(df_trap.head())

```

```python
# Plot the data
plt.plot(df_trap['time_stamp'], df_trap['percentage_left'], label="Trapezoid Profile")

# Add labels and title
plt.xlabel('Simulation Time')
plt.ylabel('SoC')
plt.title('SoC Plot for Trapezoidal Velocity Profile')
plt.legend()


plt.grid(True)
plt.grid(color='gray', linestyle='--', linewidth=0.5)

# Display the plot
plt.show()
```
